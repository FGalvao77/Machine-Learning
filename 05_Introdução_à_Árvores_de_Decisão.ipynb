# **Árvores de Decisão (Decision Trees)**

Uma `árvore de decisão` é uma representação de uma tabela de decisão sob a forma de árvore. Trata-se de uma forma alternativa de expressar as mesmas regras que são obtidas quando se constrói a tabela.

Seja a especificação do processo de cálculo do valor da conta de energia elétrica. Uma árvore de decisão para expressar tal problema seria:

image.png

É uma abordagem comportamental que usa diagramas para mapear as varias alternativas e resultados de decisões de investimento, assim como as probabilidades de ocorrerem. Baseia-se em estimativas e probabilidades associadas aos resultados de cursos de ação que competem entre si. O resultado de cada curso de ação é ponderado pela probabilidade associada a ele; o resultado ponderado é somado e o valor esperado de cada curso de ação é, então determinado. A alternativa que proporciona o valor esperado mais alto é preferível. Essencialmente, árvores de decisões são diagramas que permitem representar e avaliar problemas que envolvem decisões sequenciais, colocando em destaque os riscos e os resultados financeiros identificados nos diversos cursos de ação. Para efectuar a representação gráfica da árvore de decisão são geralmente usadas linhas para identificar a decisão (por exemplo “sim” ou “não”) e nós para identificar a as questões sobre as quais se deve decidir. Cada um dos ramos formado por linhas e nós termina numa espécie de folha que identifica a consequência mais provável da sequência de decisões tomadas.

Além da gestão, as árvores de decisão são também muito utilizadas noutras áreas, com especial destaque para a criação de algoritmos informáticos.

## **Vamos pra prática!**




**Importando as bibliotecas**

```python
# importando as funções para gerar as árvores de decisão de classificação e regressão
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor  
# importando os datasets da biblioteca do sklearn
from sklearn.datasets import load_iris, load_boston
# importando do sklearn a função "train_test_split" para dividir o conjunto de dados em treino e teste
from sklearn.model_selection import train_test_split
```

## **Árvore de Decisão para classificação**

---


  - iremos utilizar o conjunto de dados `load_iris`.


```python
# instanciando o modelo de classificação na variável "dtClas"
dtClas = DecisionTreeClassifier()
```

```python
# carregando a base de dados "load_iris" 
# e instanciando nas variáveis "X" e "y"
X, y = load_iris(return_X_y=True)

# realizando a divisão dos dados de treino e teste das variáveis "X" e "y"
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# realizando o treinamento do modelo com os dados de treino - "X_train" e "y_train"
dtClas.fit(X_train, y_train)
```

```python
# avaliando a performance do modelo
dtClas.score(X_test, y_test)
```

> Como podemos notar, o modelo foi bem performático.
  - porém isso pode não ser bom! Porquê?


O modelo pode está sofrendo de `overfitting`, em português conhecido como **sobreajuste**.

`Sobre-ajuste` ou `sobreajuste` é um termo usado em estatística para descrever quando um modelo estatístico se ajusta muito bem ao conjunto de dados anteriormente observado, mas se mostra ineficaz para prever novos resultados. É comum que a amostra apresente desvios causados por erros de medição ou fatores aleatórios.

- fonte:
  - Wikipedia: https://bit.ly/2TvcFQM

E em contraoposição temos o `underfitting`, em tradução livre para o português, **sub-ajustado.**

Underfitting é uma tradução para o inglês de sub-ajustado. Ou seja, nosso modelo não conseguiu aprender suficiente sobre os dados. O underfitting leva à um erro elevado tanto nos dados de treino quando nos dados de teste.

- fonte:
  - ABRACD: https://bit.ly/3znlOLL

**E como resolver a ocorrência do** `overfitting`**?**

**Causas e soluções**

O _overfitting_ tem algumas causas principais, que podem direcionar a solução do problema:

- `Algoritmo muito complexo para os dados`: podemos simplificar nosso modelo escolhendo um algoritmo mais simples, com menos parâmetros, caso seja possível. Isso reduz as chances do modelo sofrer _overfitting_.

- `Poucos dados de treinamento`: talvez seja necessário coletar mais dados para treinar o modelo.

- `Ruídos nos dados de treinamento`: Caso exista algum tipo de ruído (valores extremos ou até mesmo valores incorretos nos dados), pode ser que o modelo aprenda sobre ele, levando ao overfitting. Caberia um pré-processamento adequado para tratar essa interferência.

Porém, se formos muito rígidos nos tratamentos acima, podemos ir para o outro extremo, o _underfitting_.

- `Algoritmo inadequado, pouco poderoso para os dados`: aqui podemos amplificar o poder do nosso algoritmo escolhendo outro com mais parâmetros para solucionar o underfitting.

- `Características não representativas`: neste caso, pode ser que as características que estamos utilizando para treinar o modelo não sejam representativas (não tenham relação entre si ou não sejam importantes para o modelo).

- `Modelo com muitos parâmetros de restrição`: o modelo torna-se inflexível, restrito, e não consegue se ajustar de forma adequada aos dados.

## **Árvore de Decisão para classificação**

---


  - iremos utilizar o conjunto de dados `load_iris`.

```python
# importando a função para gerar a árvore de decisão de classificação
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor  
```

```python
# importando o dataset da biblioteca do sklearn
from sklearn.datasets import load_iris
```

```python
# iremos instanciar dois modelos de classificação
# e passando como argumento no primeiro modelo "max_depth=5" e, no segundo "max_depth=3"
# isso significa que queremos, 5 nós no primeiro e três nós no segundo modelo
dtClas_1 = DecisionTreeClassifier(max_depth=5)
dtClas_2 = DecisionTreeClassifier(max_depth=3)
```

```python
# carregando a base de dados "load_iris" 
# e instanciando nas variáveis "X" e "y"
X, y = load_iris(return_X_y=True)

# realizando a divisão dos dados de treino e teste das variáveis "X" e "y"
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# realizando o treinamento do modelo com os dados de treino - "X_train" e "y_train"
print(dtClas_1.fit(X_train, y_train))
print('')
print(dtClas_2.fit(X_train, y_train))
```

```python
# avaliando a performance do modelo
print('Árvore com 5 nós: ', dtClas_1.score(X_test, y_test))
print('Árvore com 3 nós: ', dtClas_2.score(X_test, y_test))
```

> Perceba que, a árvore mais simples com 3 nós apresentou uma performance bem melhor que com 5 nós.
  - portanto, podemos concluir que a aumentar a quantidade de `nós` e a complexidade do modelo não é garantia de um modelo ideal, já que o mesmo poderá estar sofrendo de `overfitting`.

## **Árvore de Decisão para classificação**

---


  - iremos utilizar o conjunto de dados `load_breast_cancer`.

```python
# importando as bibliotecas e o conjunto de dados
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_breast_cancer
```

```python
# instanciando o modelo
dtClas = DecisionTreeClassifier()
```

```python
# realizando a leitura do dataset e instanciando nas variáveis "X" e "y"
X, y = load_breast_cancer(return_X_y=True)
```

```python
# treinando o modelo com os dados das variáveis - "X" e "y"
dtClas.fit(X, y)
```

```python
# visualizando a quantidade de linhas e colunas
X.shape
```

> Note que, temos 30 colunas! Mas será que todas essas colunas tem importância na construção da nossa `árvore de decisão`.
  - pois bem, para saber essa importância temos a função `.features_importances_` que retorna quanto cada coluna tem influência em nossa árvore.

**Vamos aplicar a função e descobrir a importância das features.**

```python
# visualizando a importância das "features" na geração da árvore de decisão
dtClas.feature_importances_
```

> Como podemos ver, temos o retorno de um array com 30 elementos que mostra a importância de cada feature para construção do modelo de `árvore de decisão`, algumas colunas tem valores bem baixo e até zero. 
  - podemos eliminar as features de importância baixa e de valor zero. Porém isso dever ser feito recursivamente - uma à uma e, depois rodamos o modelo novamente e avaliamos sua performance. 
  
  **Porque não podemos eleminar de uma só vez as features de baixa importância?** 
    
  - pois alguma coluna pode estar `correlacionada` com outra e, se eliminarmos todas as features de baixa importância podemos prejudicar a feature que apresenta um valor de importância razoável.
